```python
import os
import json
from urllib.request import urlretrieve
from tqdm import tqdm
import pathlib
from shutil import copyfile
```

```python
labelbox_path = os.path.abspath('../data/train/export-2019-10-23T15_28_50.623Z.json')
image_src_dir = '/run/media/soedomoto/SOEDOMOTO/DATA/PROJECT/ICAS2019/skynet-data/data/indonesia/images/'
train_images_dir = os.path.abspath('../data/train/images/')
train_masks_dir = os.path.abspath('../data/train/masks/')
```

```python
with open(labelbox_path, 'r') as f:
    labelbox_json = json.load(f)
```

```python
labelbox_json
```

```python
image_mask_pair = []
for label in labelbox_json:
    image_name = label['External ID']
    image_mask_paddy_name = None
    for obj in label['Label']['objects']:
        if obj['value'] == 'paddy':
            image_mask_paddy_name = obj['instanceURI']
            break
            
    if image_name and image_mask_paddy_name:
        image_mask_pair.append((image_name, image_mask_paddy_name))
```

```python
image_mask_pair
```

```python
pathlib.Path(train_images_dir).mkdir(parents=True, exist_ok=True)
pathlib.Path(train_masks_dir).mkdir(parents=True, exist_ok=True)

for i in tqdm(range(len(image_mask_pair))):
    img_name, mask = image_mask_pair[i]
    copyfile(os.path.join(image_src_dir, img_name), os.path.join(train_images_dir, img_name))
    
for i in tqdm(range(len(image_mask_pair))):
    img_name, mask = image_mask_pair[i]
    urlretrieve(mask, os.path.join(train_masks_dir, img_name.split('.')[0] + '.png'))
```
